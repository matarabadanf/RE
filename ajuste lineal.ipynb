```python
import numpy as np
import pandas as pd
import statsmodels.api as sm
import math

y = [0.0, 3.0, 5.0, 6, 10.0, 13.0, 15.0, 20.0, 25.0, 30.0, 35.0]
x = [0.0, 6.0, 10.0, 12, 20.0, 26.0, 30.0, 40.0, 50.0, 60.0, 70.0]

y = np.array(y)
t = np.array(x)

t = sm.add_constant(t, prepend=False)

model = sm.OLS(y,t)
result = model.fit()
result.summary()
```
